# Задача 1
Написать функцию на вход которой подается строка, состоящая из латинских букв.
Функция должна вернуть количество гласных букв (a, e, i, o, u) в этой строке. Написать тесты для кода

```python
def count_vowels(string):
    vowels = set('aeiou')
    return sum(1 for char in string.lower() if char in vowels)

# Тесты
assert count_vowels('hbvv41kugdsai') == 3
assert count_vowels('oooooh my nword') == 6
assert count_vowels('AEIOU') == 5
assert count_vowels('bcdfg') == 0
print('✅ Задача 1 пройдена')
```

# Задача 2
Написать функцию на вход, которой подается строка. Функция должна вернуть true, если
каждый символ в строке встречается только 1 раз, иначе должна вернуть false. Написать тесты для кода

```python
def is_unique(string):
    return len(set(string)) == len(string)

# Тесты
assert is_unique("qwerty123") == True
assert is_unique("SpatHochu") == True
assert is_unique("TruLaLa") == False
assert is_unique("aA") == True  # регистр учитывается
assert is_unique("") == True
print('✅ Задача 2 пройдена')
```

# Задача 3
Написать функцию, которая принимает положительное число и возвращает количество
бит равных 1 в этом числе. Написать тесты для кода

```python
def count_set_bits(n):
    return bin(n).count('1')

# Тесты
assert count_set_bits(10) == 2  # 1010
assert count_set_bits(0) == 0
assert count_set_bits(1) == 1
assert count_set_bits(7) == 3  # 111
assert count_set_bits(15) == 4 # 1111
print('✅ Задача 3 пройдена')
```

# Задача 4
Написать функцию, которая принимает положительное число. Функция должна вернуть то,
сколько раз необходимо перемножать цифры числа или результат перемножения, чтобы
получилось число состоящее из одной цифры.
Например, для входного числа:
· 39 функция должна вернуть 3, так как 3*9=27 => 2*7=14 => 1*4=4
· 4 функция должна вернуть 0, так как число уже состоит из одной цифры
· 999 функция должна вернуть 4, так как 9*9*9=729 => 7*2*9=126 => 1*2*6=12 =>
1*2=2. Написать тесты для кода

```python
def persistence(n):
    steps = 0
    while n >= 10:
        product = 1
        for digit in str(n):
            product *= int(digit)
        n = product
        steps += 1
    return steps

# Тесты
assert persistence(39) == 3
assert persistence(4) == 0
assert persistence(999) == 4
assert persistence(10) == 1  # 1*0 = 0
assert persistence(25) == 2  # 2*5=10 → 1*0=0
print('✅ Задача 4 пройдена')
```

# Задача 5
Написать функцию, которая принимает два целочисленных вектора одинаковой длины и
возвращает среднеквадратическое отклонение двух векторов. Написать тесты для кода

```python
def mse(pred, true):
    if len(pred) != len(true):
        raise ValueError("Vectors must have the same length")
    return sum((p - t) ** 2 for p, t in zip(pred, true)) / len(pred)

# Тесты
assert mse([1, 2, 3], [1, 2, 3]) == 0.0
assert mse([1, 2], [2, 3]) == 1.0
assert abs(mse([0, 0], [1, 1]) - 1.0) < 1e-9
assert mse([3], [5]) == 4.0
print('✅ Задача 5 пройдена')
```

# Задача 6
Написать функцию, принимающая целое положительное число. Функция должна вернуть
строку вида “(n1**p1)(n2**p2)…(nk**pk)” представляющая разложение числа на простые
множители (если pi == 1, то выводить только ni).
Например, для числа 86240 функция должна вернуть “(2**5)(5)(7**2)(11)”. Написать тесты для кода

```python
def prime_factorization(n):
    if n <= 1:
        return ""
    factors = {}
    d = 2
    while d * d <= n:
        while n % d == 0:
            factors[d] = factors.get(d, 0) + 1
            n //= d
        d += 1
    if n > 1:
        factors[n] = factors.get(n, 0) + 1
    
    parts = []
    for prime in sorted(factors.keys()):
        exp = factors[prime]
        if exp == 1:
            parts.append(f"({prime})")
        else:
            parts.append(f"({prime}**{exp})")
    return ''.join(parts)

# Тесты
assert prime_factorization(86240) == "(2**5)(5)(7**2)(11)"
assert prime_factorization(1) == ""
assert prime_factorization(2) == "(2)"
assert prime_factorization(12) == "(2**2)(3)"
assert prime_factorization(17) == "(17)"
print('✅ Задача 6 пройдена')
```

# Задача 7
Написать функцию, принимающая целое число n, задающее количество кубиков. Функция
должна определить, можно ли из данного кол-ва кубиков построить пирамиду, то есть
можно ли представить число n как 1^2+2^2+3^2+…+k^2. Если можно, то функция должна
вернуть k, иначе строку “It is impossible”. Написать тесты для кода

```python
def pyramid(n):
    k = 1
    total = 0
    while total < n:
        total = k * (k + 1) * (2 * k + 1) // 6
        if total == n:
            return k
        k += 1
    return "It is impossible"

# Тесты
assert pyramid(1) == 1
assert pyramid(5) == "It is impossible"
assert pyramid(14) == 3  # 1+4+9=14
assert pyramid(30) == 4  # 1+4+9+16=30
assert pyramid(0) == "It is impossible"
print('✅ Задача 7 пройдена')
```

# Задача 8
Функция принимает на вход положительное число и определяет является ли оно сбалансированным, т.е. сумма цифр до средних равна сумме цифр после. Средними в случае нечетного числа цифр считать одну цифру, в случае четного - две средних. Написать тесты для кода

```python
def is_balanced_number(n):
    s = str(n)
    L = len(s)
    if L % 2 == 1:
        mid = L // 2
        left_part = s[:mid]
        right_part = s[mid+1:]
    else:
        mid = L // 2
        left_part = s[:mid-1]
        right_part = s[mid+1:]
    left_sum = sum(int(d) for d in left_part)
    right_sum = sum(int(d) for d in right_part)
    return left_sum == right_sum

# Тесты
assert is_balanced_number(12321) == True   # 1+2 == 2+1
assert is_balanced_number(1221) == True    # 1 == 1
assert is_balanced_number(1234) == False   # 1 != 4
assert is_balanced_number(5) == True       # нет частей → 0 == 0
assert is_balanced_number(123) == True     # 1 == 3? → Нет! Ой...
# Исправим: 123 → средняя '2', лево='1', право='3' → 1 != 3 → False
assert is_balanced_number(123) == False
assert is_balanced_number(121) == True     # 1 == 1
print('✅ Задача 8 пройдена')
```
